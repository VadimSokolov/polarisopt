```python
model_folder = '/home/vsokolov/Austin/'
json_fn = "CampoTimingChoiceModel.json"
polaris_file = model_folder + json_fn
key = 'Activity_Timing_Choice_Model'
out_config_fn = '../scratch/config_morris_timing.json'
in_file = "../scratch/focus-parameters.json"
```

```python
pd = json.load(open(in_file,'r'))
```

```python
maxval = {p:0 for p in pd}
maxval
```

```python
res = []
for p in pd:
    v = pd[p]
    # l = (v-3*abs(v))
    l=-1
    u = (v+2*abs(v))
    if p in maxval:
        u = maxval[p]
    d = {"name": p,"type": "float","min": l,"max": u,"size": 1}
    res.append(d)
d = {json_fn:res}
```
