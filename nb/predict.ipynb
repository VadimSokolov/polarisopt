```python
import pandas as pd
import json
import glob
import os
import h5py
import numpy as np
import matplotlib.pyplot as plt
import datetime
```

```python
with open("/home/vsokolov/Austin/morris/config_morris_constants.json",'r') as fh:
    config = json.load(fh)
polaris_json_file = list(config.keys())[0]
vars = [item['name'] for item in config[polaris_json_file]]
polaris_json_file
```

```python
f'/projects/vsokolov/austin-morris/Sim0/{polaris_json_file}'
```

```python
d = []
for sim in glob.glob(f'/projects/vsokolov/austin-morris/**/{polaris_json_file}'):    
    with open(sim) as fh:
        data = json.load(fh)
    vals = [data['Activity_Generation_Model'][item] for item in vars]
    vals= vals+[os.path.dirname(sim)]
    d.append(vals)

```

```python
df = pd.DataFrame(d, columns=vars+['folder'])
df.to_csv('vals-folder.csv', index=False)
```

```python
df.iloc[0]['folder'].split('/')[-1]
```

```python
for item in df['folder']: 
    simname = item.split('/')[-1]
    print(simname)
    f = h5py.File(f"{item}/Austin_iteration_10/Austin-Result.h5", 'r')
    a = f['link_moe']['link_in_flow_ratio'][:]
    y = np.mean(a, axis=1)
    plt.plot(y)
    plt.savefig(f'fig/{simname}_link_in_flow_ratio.png')
    plt.close()
```
