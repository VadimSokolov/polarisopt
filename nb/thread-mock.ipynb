```python
from concurrent import futures
from itertools import repeat
from random import randint
from time import sleep
```

```python
def sim(x):
    print(f'Processing {x}')
    sleep(randint(1, 3))
    with open(f'thread-mock-{x}.txt', 'a') as f:
        f.write(f'{x}\n')
    print(f'Finished {x}')
    return x*x
```

```python
n = 20
with futures.ThreadPoolExecutor(5) as executor:
    result = executor.map(sim, range(n))
    for item in result:
        print(item)
```
