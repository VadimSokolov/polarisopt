```python
import sqlite3,h5py
import glob,os
import pandas as pd
```

```python
ef = '/home/vsokolov/projects/timing-austin/experiments/'
st = pd.DataFrame(columns=['experiment', 'status'])
for s in glob.glob(f'{ef}/Sim*'):
    experiment = s.split('/')[-1].strip('Sim')
    status = 0
    for it in glob.glob(f'{s}/Austin_iteration*'):
        if os.path.exists(f'{it}/finished'):
            status+=1
    st.loc[len(st)] = [experiment, status]
    
```

```python
with pd.option_context('display.max_rows', None):
    print(st)
```
